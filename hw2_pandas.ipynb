#### Задание 1. Соберите всю информацию о студентах в одну таблицу df. В получившейся таблице должна быть информация и оценки всех студентов из всех групп. Напечатайте несколько строк таблицы для демонстрации результата.¶

```python
!wget https://github.com/hse-ds/iad-intro-ds/raw/master/2023/homeworks/Data.zip #скачала файл с данными
!unzip /content/Data.zip #извлекла данные из файла

df = pd.DataFrame()
for i in range(0, 10): #благодаря циклу читаем каждый файл с одним индексом, пересекаем две таблицы, получая единую с инфой и оценками, потом склеиваем полученную таблицу с сформированной до того, проходя по всем файлам (10 кругов) получаем полный датасет
    students_info = pd.read_csv(f'/content/Data/Students_info_{i}.csv')
    students_marks = pd.read_csv(f'/content/Data/Students_marks_{i}.csv')

    merged_data = pd.merge(students_info, students_marks, on='index') #пересечение по столбцу с индексами
    df = pd.concat([df, merged_data], ignore_index=True) #склеивание таблиц с игнорированием индексов)
df
```

#### Задание 13. Сколько студентов успешно сдали экзамен по математике?

Создайте новый столбец в таблице df под названием Math_PassStatus и запишите в него F, если студент не сдал экзамен по математике (балл за экзамен < passmark), и P иначе.

Посчитайте количество студентов, сдавших и не сдавших экзамен по математике.

Сделайте аналогичные шаги для экзаменов по чтению и письму.

```python
#математика

def sortt(score):
  if score < passmark:
    return 'F'
  else:
    return 'P'

df['Math_PassStatus'] = df['math score'].apply(sortt)

print(f"Количество студентов, сдавшие экзамен по математике = {np.sum(df['Math_PassStatus'] == 'P')}")
print(f"Количество студентов, не сдавшие экзамен по математике = {np.sum(df['Math_PassStatus'] == 'F')}")
```

```python
#чтение

def sortt(score):
  if score < passmark:
    return 'F'
  else:
    return 'P'

df['Read_PassStatus'] = df['reading score'].apply(sortt)

print(f"Количество студентов, сдавшие экзамен по чтению = {np.sum(df['Read_PassStatus'] == 'P')}")
print(f"Количество студентов, не сдавшие экзамен по чтению = {np.sum(df['Read_PassStatus'] == 'F')}")
```

```python
#письмо

def sortt(score):
  if score < passmark:
    return 'F'
  else:
    return 'P'

df['Write_PassStatus'] = df['writing score'].apply(sortt)

print(f"Количество студентов, сдавшие экзамен по письму = {np.sum(df['Write_PassStatus'] == 'P')}")
print(f"Количество студентов, не сдавшие экзамен по письму = {np.sum(df['Write_PassStatus'] == 'F')}")
```

#### Задание 14. Сколько студентов успешно сдали все экзамены?

Создайте столбец OverAll_PassStatus и запишите в него для каждого студента 'F', если студент не сдал хотя бы один из трех экзаменов, а иначе 'P'.

Посчитайте количество студентов, которые сдали все экзамены.

```python
import numpy as np

def select(ind):
  if ind['math score'] >= passmark and ind['reading score'] >= passmark and ind['writing score'] >= passmark:
    return 'P'
  else:
    return 'F'

df['OverAll_PassStatus'] = df.apply(select, axis = 1)

print(f"Количество студентов, которые сдали все экзамены = {np.sum(df['OverAll_PassStatus'] == 'P')}")
```

### Система перевода баллов в оценки
####    больше 90 = A
####      80-90 = B
####      70-80 = C
####      60-70 = D
####      50-60 = E
####    меньше 50 = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

**В случае, если средний балл попадает на границу между оценками (т.е. равен ровно 60, 70 или 80 баллов), вы можете интерпретировать условие на своё усмотрение (т.е. можете поставить за 60 баллов оценку D, а можете - E).**

```python
import numpy as np

def GetGrade(average_mark):
    if average_mark > 90:
      return 'A'
    elif 90 >= average_mark > 80:
      return 'B'
    elif 80 >= average_mark > 70:
      return 'C'
    elif 70 >= average_mark > 60:
      return 'D'
    elif 60 >= average_mark > 50:
      return 'E'
    else:
      return 'F (Fail)'


for i in range(len(df)):
  df.loc[i, 'Grade'] = np.mean([(df.loc[i, 'math score']), (df.loc[i, 'reading score']), (df.loc[i, 'writing score'])])

df['Grade'] = df['Grade'].apply(GetGrade)

print(f"Количество студентов, получившие оценку A = {np.sum(df['Grade'] == 'A')}")
print(f"Количество студентов, получившие оценку B = {np.sum(df['Grade'] == 'B')}")
print(f"Количество студентов, получившие оценку C = {np.sum(df['Grade'] == 'C')}")
print(f"Количество студентов, получившие оценку D = {np.sum(df['Grade'] == 'D')}")
print(f"Количество студентов, получившие оценку E = {np.sum(df['Grade'] == 'E')}")
print(f"Количество студентов, получившие оценку F (Fail) = {np.sum(df['Grade'] == 'F (Fail)')}")
```
